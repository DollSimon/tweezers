```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
from ixo.file import get_similar_files
from tweezer.io import TxtSourceMpi
from tweezer import TweezerData
from tweezer.plot.psd import PsdFitPlot
```

### Batch thermal calibration

Run the thermal calibration in batch on all matching files in the given directory.

```python
# find files
path = '/Users/christophehrlich/Documents/TranscriptionData/D3/refit'
files = get_similar_files(path, foldersApart=1, categories=[['', '.txt'], ['PSD_', '.txt'], ['TS_', '.txt']])

for file in files:
    print('Processing: ' + str(file[0]))
    
    # create data source for file
    source = TxtSourceMpi(data=file[0], psd=file[1], ts=file[2])
    # create data object
    data = TweezerData(source)
    data.compute_psd().fit_psd(maxF=3000)
    # write stuff back to data file
    source.write_data(data)
```

### Single calibration

The thermal calibration is performed by computing the power spectrum of random bead fluctuations and fitting the therotically predicted Lorentzian to it.

```python
# get files
path = '/Users/christophehrlich/Documents/TranscriptionData/D3/refit'
files = get_similar_files(path, foldersApart=1, categories=[['', '.txt'], ['PSD_', '.txt'], ['TS_', '.txt']])
if not files or not files[0]:
    raise ValueError('No files found')

# create data source
source = TxtSourceMpi(data=files[0][0], psd=files[0][1], ts=files[0][2])
# create data object
data = TweezerData(source)
# compute PSD and fit it
data.compute_psd().fit_psd(maxF=3000)

# plot
PsdFitPlot(data)
```
