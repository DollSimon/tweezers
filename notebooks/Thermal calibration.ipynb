```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
from ixo.file import get_files, get_similar_files
from tweezers.io import TxtSourceMpi
from tweezers import TweezerData
from tweezers.plot.psd import PsdFitPlot
import re
from physics.viscosity import dynamic_viscosity_water_glycerol
```

### Batch thermal calibration

Run the thermal calibration in batch on all matching files in the given directory. A nicer way to find the correct files could be required here.

```python
# get all files
# path = '/Volumes/dualtrapgrill/vero/FINAL_CORRECTION/New_Fitting_LS_25_weigthedFit/'
path = '/Users/christophehrlich/Documents/TranscriptionData/D3/refit'
files = get_files(path, suffix='.txt', recursive=True)

# sort them
dataFiles = []
tsFiles = []
psdFiles = []
for file in files:
    if file.name[0].isdigit():
        dataFiles.append(file)
    elif file.name.startswith('TS_'):
        tsFiles.append(file)
    elif file.name.startswith('PSD_'):
        psdFiles.append(file)

# find matching pairs in data and ts
tsRegex = re.compile('^TS_(?P<name>.*?)\.txt$')
dataRegex = re.compile('^(?P<name>.*?)(_zoom.*)?\.txt$')
matchedFiles = []
for tsFile in tsFiles:
    tsName = tsRegex.match(tsFile.name).groupdict()['name']
    for dataFile in dataFiles:
        dataName = dataRegex.match(dataFile.name).groupdict()['name']
        # check for same name and similar parent folders
        if tsName == dataName and tsFile.parents[1] == dataFile.parents[1]:
            matchedFiles.append([dataFile, tsFile])
```

```python
water = 0.9
glycerol = 0.1
temperature = 25

for file in matchedFiles:
    print('Processing: ' + str(file[0]))
    
    # create data source for file
    source = TxtSourceMpi(data=file[0], ts=file[1])
    # create data object
    data = TweezerData(source)
    data.meta['temperature'] = temperature
    data.meta['viscosity'] = 1e-6 * dynamic_viscosity_water_glycerol(water, glycerol, temperature)
    data.compute_psd().fit_psd(maxF=3000)
    # write stuff back to data file
    source.write_data(data)

print('Done')
```

### Single calibration

The thermal calibration is performed by computing the power spectrum of random bead fluctuations and fitting the therotically predicted Lorentzian to it.

```python
# get files
path = '/Users/christophehrlich/Documents/TranscriptionData/D3/refit'
files = get_similar_files(path, foldersApart=1, categories=[['', '.txt'], ['PSD_', '.txt'], ['TS_', '.txt']])
if not files or not files[0]:
    raise ValueError('No files found')

# create data source
source = TxtSourceMpi(data=files[0][0], psd=files[0][1], ts=files[0][2])
# create data object
data = TweezerData(source)
# compute PSD and fit it
data.compute_psd().fit_psd(maxF=3000)

# plot
PsdFitPlot(data)
```
